```python
# Importing necessary libraries
import pandas as pd
import glob
from google.colab import files
print("Libraries imported – ready to start.")
```

```python
# Uploading all 12 cleaned Booking Excel files
uploaded = files.upload()
print("All files uploaded.")
```

```python
# Get the list of uploaded Booking files (sorted alphabetically – matches your screenshot)
files_list = sorted(glob.glob('*_Booking_Cleaned.xlsx'))

# Print to confirm the order
print("Your uploaded Booking files (in order):")
for i, file in enumerate(files_list, 1):
    print(f"{i}. {file}")

# Define the correct hotel names in the SAME order as the files above
hotel_names = [
    'Cinnamon Citadel Kandy',
    'Habarana Village by Cinnamon',
    'Hikka Tranz by Cinnamon',
    'Kandy Mist by Cinnamon',
    'Trinco Blu by Cinnamon',
    'Cinnamon Bey Beruwala',
    'Cinnamon Grand Colombo',
    'Cinnamon Lakeside Colombo',
    'Cinnamon Life Colombo',
    'Cinnamon Lodge Habarana',
    'Cinnamon Red Colombo',
    'Cinnamon Wild Yala'
]

print(f"\nNumber of files: {len(files_list)}")
print(f"Number of hotel names defined: {len(hotel_names)}")

if len(files_list) == len(hotel_names):
    print("Order matches – ready to merge!")
else:
    print("Mismatch – check file order or hotel list.")
```

```python
# Creating an empty DataFrame to store the merged data
booking_master = pd.DataFrame()

# Loop through each file and hotel name
for file, hotel in zip(files_list, hotel_names):
    # Read the Excel file
    df = pd.read_excel(file)

    # Add the hotel name to every row in this file
    df['Hotel/Resort'] = hotel

    # Add platform identifier
    df['Platform'] = 'Booking.com'

    # Append to the master DataFrame
    booking_master = pd.concat([booking_master, df], ignore_index=True)

    # Print progress
    print(f"Added {len(df)} reviews from {hotel}")

# Final check
print(f"\nMerging complete!")
print(f"Total reviews in booking_master: {len(booking_master)}")
print(f"Total hotels: {booking_master['Hotel/Resort'].nunique()}")
```

```python
# Check total reviews
print(f"Total reviews: {len(booking_master)}")

# Check how many reviews per hotel
print("\nReviews per hotel:")
print(booking_master['Hotel/Resort'].value_counts().sort_index())

# Check if 'Hotel/Resort' is filled for all rows
print(f"\nMissing Hotel/Resort values: {booking_master['Hotel/Resort'].isnull().sum()}")

# Show a few sample rows from different hotels
print("\nSample rows from different hotels:")
display(booking_master.groupby('Hotel/Resort').head(1)[['Hotel/Resort', 'Platform', 'ReviewDate', 'Rating']].sort_values('Hotel/Resort'))
```

```python
# Dropping the below column because its values are in codes so it is harder to understand
columns_to_drop = [

    'HotelRating'
]

booking_master = booking_master.drop(columns=columns_to_drop, errors='ignore')
```

```python
# Re-naming/Standardizing the columns
booking_master = booking_master.rename(columns={

    'AverageStaffRating': 'StaffRating',
    'AverageFacilitiesRating': 'FacilitiesRating',
    'AverageCleanlinessRating': 'CleanlinessRating',
    'AverageComfortRating': 'ComfortRating',
    'AverageValueforMoneyRating': 'ValueforMoneyRating',
    'AverageLocationRating': 'LocationRating',
    'AverageFreeWifiRating': 'FreeWifiRating',
    'RoomInformation': 'RoomtipText',
    'TravelerType': 'TripType',
    'Rating': 'HotelRating'

})
```

```python
# Listing of all rating columns (after your renaming)
rating_columns = [
    'HotelRating',
    'StaffRating',
    'FacilitiesRating',
    'CleanlinessRating',
    'ComfortRating',
    'ValueforMoneyRating',
    'LocationRating',
    'FreeWifiRating',

]

# Verifying they exist
missing = [col for col in rating_columns if col not in booking_master.columns]
if missing:
    print(f"Warning: Missing columns: {missing}")
else:
    print("All rating columns found – ready to normalize.")
```

```python
# Divide all selected rating columns by 2 (1-10 → 1-5 scale)
booking_master[rating_columns] = booking_master[rating_columns] / 2

print("All Booking.com ratings successfully divided by 2.")
```

```python
# Round to 1 decimal for clean, comparable values
booking_master[rating_columns] = booking_master[rating_columns].round(1)

print("Ratings rounded to 1 decimal place.")
```

```python
# Show average ratings after normalization (should now be in 3.5–4.8 range)
print("Average Ratings After Normalization (1-5 scale):")
print(booking_master[rating_columns].mean().round(2).sort_values(ascending=False))

# Sample a few rows to confirm
print("\nSample normalized ratings:")
display(booking_master[rating_columns].head(5))
```

```python
# Save the final normalized version
booking_master.to_excel('Booking_Master_Normalized_Ratings.xlsx', index=False)
print("Normalized Booking master saved.")

from google.colab import files
files.download('Booking_Master_Normalized_Ratings.xlsx')
```

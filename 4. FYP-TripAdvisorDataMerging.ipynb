```python
# Importing necessary libraries
import pandas as pd
import glob
from google.colab import files
print("Libraries imported ‚Äì ready to start.")
```

```python
# Upload all 12 cleaned TripAdvisor Excel files
uploaded = files.upload()  # Select all 12 files ending with _TA_Cleaned.xlsx
print("TripAdvisor files uploaded.")
```

```python
# Get sorted files (matches your screenshot alphabetical order)
files_list_ta = sorted(glob.glob('*_TA_Cleaned.xlsx'))

print("TripAdvisor files (exact order from screenshot):")
for i, file in enumerate(files_list_ta, 1):
    print(f"{i}. {file}")

# Hotel names EXACTLY matching file order from your image
hotel_names_ta = [
    'Cinnamon Grand Colombo',           # CinnGrand_Colombo_TA_Cleaned.xlsx
    'Cinnamon Lakeside Colombo',        # CinnLakeside_Colombo_TA_Cleaned.xlsx
    'Cinnamon Red Colombo',             # CinnRed_Colombo_TA_Cleaned.xlsx
    'Cinnamon Life Colombo',            # CinnLife_Colombo_TA_Cleaned.xlsx
    'Kandy Mist by Cinnamon',           # Cinn_KandyMyst_TA_Cleaned.xlsx
    'Cinnamon Bey Beruwala',            # CinnBey_Beruwala_TA_Cleaned.xlsx
    'Hikka Tranz by Cinnamon',          # CinnHikkaTranz_TA_Cleaned.xlsx
    'Cinnamon Lodge Habarana',          # CinnLodge_Habarana_TA_Cleaned.xlsx
    'Cinnamon Citadel Kandy',           # CinnCitadelKandy_TA_Cleaned.xlsx
    'Cinnamon Wild Yala',               # CinnWild_Yala_TA_Cleaned.xlsx
    'Habarana Village by Cinnamon',     # Cinn_HabaranaVillage_TA_Cleaned.xlsx
    'Trinco Blu by Cinnamon'            # CinnTrincoBlu_TA_Cleaned.xlsx
]

print(f"\nFiles found: {len(files_list_ta)} | Hotels defined: {len(hotel_names_ta)}")
print("Status:", "‚úÖ READY" if len(files_list_ta) == 12 else "‚ùå CHECK FILES")
```

```python
# Initialize master DataFrame
tripadvisor_master = pd.DataFrame()

# Loop through files and assign hotel names
for file, hotel in zip(files_list_ta, hotel_names_ta):
    df = pd.read_excel(file)

    # Add Hotel/Resort column to EVERY row
    df['Hotel/Resort'] = hotel

    # Add Platform column
    df['Platform'] = 'TripAdvisor'

    # Optional: Standardize some column names for consistency with Booking
    df = df.rename(columns={
        'Language': 'ReviewLanguage',
        'RatingPublishedDate': 'ReviewDate',
        'ReviewText': 'LikedText'  # Unified text field for NLP
    })

    # Append to master
    tripadvisor_master = pd.concat([tripadvisor_master, df], ignore_index=True)

    print(f"‚úÖ {hotel}: {len(df)} reviews added")

print(f"\nüéâ TRIPADVISOR MASTER COMPLETE!")
print(f"Total reviews: {len(tripadvisor_master)}")
print(f"Unique hotels: {tripadvisor_master['Hotel/Resort'].nunique()}")
print("Columns:", tripadvisor_master.columns.tolist())
```

```python
# Save the complete TripAdvisor master
tripadvisor_master.to_excel('TripAdvisor_Master_Final.xlsx', index=False)
print("TripAdvisor Master saved successfully.")

from google.colab import files
files.download('TripAdvisor_Master_Final.xlsx')
```

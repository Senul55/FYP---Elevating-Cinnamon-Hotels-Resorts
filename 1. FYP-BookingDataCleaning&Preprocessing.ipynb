```python
# Printing a list of all the columns in the dataset
print("All columns:")
print(CinnGrand_Colombo_BK.columns.tolist())

print("\nData types and info:")
CinnGrand_Colombo_BK.info()
```

```python
# Checking the missing percentage of data in each column
missing = CinnGrand_Colombo_BK.isnull().sum()
print("Missing values per column:")
print(missing[missing > 0])

print("\nMissing percentage:")
print((missing / len(CinnGrand_Colombo_BK) * 100).round(2))
```

```python
# Checking if there are any duplicate columns
duplicates = CinnGrand_Colombo_BK.duplicated().sum()
print(f"Total fully duplicate rows: {duplicates}")

if duplicates > 0:
    print("Example duplicates:")
    display(CinnGrand_Colombo_BK[CinnGrand_Colombo_BK.duplicated()].head())
```

```python
# Checking if the columns are properly dropped
CinnGrand_Colombo_BK.head()
```

```python
# Re-checking all the column names, non-null count and the type of data
print("Data types and info:")
CinnGrand_Colombo_BK.info()
```
